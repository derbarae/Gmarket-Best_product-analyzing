### 2. 연습문제 풀이

<div class="alert alert-block alert-warning">
<font color="blue" size="4em">연습문제1</font><br>
전체 베스트상품(ALL 메인 카테고리)에서 판매자별 베스트상품 갯수 출력해보기
</div>

```python
sql1 = """SELECT A.provider, count(*) FROM items A
INNER JOIN ranking B ON A.item_code = B.item_code
WHERE B.main_category = 'ALL'
GROUP BY provider
"""
df1 = pd.read_sql(sql1, db)
df1.head()

#     여러 요소 선택시 콤마 까먹지말고!
```

<div class="alert alert-block alert-warning">
<font color="blue" size="4em">연습문제2</font><br>
메인 카테고리가 패션의류인 서브 카테고리 포함, 패션의류 전체 베스트상품에서 판매자별 베스트상품 갯수가 5이상인 판매자와 베스트상품 갯수 출력해보기
</div>

```python
sql2 = """SELECT A.provider, count(*) FROM items A
INNER JOIN ranking B ON A.item_code = B.item_code
WHERE B.main_category = '패션의류'
GROUP BY A.provider HAVING count(*) >= 5
"""
df2 = pd.read_sql(sql2, db)
df2
```

```python
sql2 = """SELECT A.provider, count(*) FROM items A
INNER JOIN ranking B ON A.item_code = B.item_code
WHERE B.main_category = '패션의류'
GROUP BY A.provider HAVING count(*) >= 5
ORDER BY count(*) DESC
"""
# ORDER BY 추가
df2 = pd.read_sql(sql2, db)
df2
```

<div class="alert alert-block alert-warning">
<font color="blue" size="4em">연습문제3</font><br>
메인 카테고리가 신발/잡화인 서브 카테고리 포함, 전체 베스트상품에서 판매자별 베스트상품 갯수가 10이상인 판매자와 베스트상품 갯수를 베스트상품 갯수 순으로 출력해보기
</div>

```python
sql3 = """SELECT A.provider, count(*) FROM items A
INNER JOIN ranking B ON A.item_code = B.item_code
WHERE B.main_category ='신발/잡화'
GROUP BY A.provider HAVING count(*) >= 10
ORDER BY count(*) DESC
"""
df3 = pd.read_sql(sql3, db)
df3
```

<div class="alert alert-block alert-warning">
<font color="blue" size="4em">연습문제4</font><br>
메인 카테고리가 화장품/헤어인 서브 카테고리 포함, 전체 베스트상품의 평균, 최대, 최소 가격 출력해보기
</div>


```python
sql4 = """SELECT avg(A.dis_price), max(A.dis_price), min(A.dis_price) FROM items A
INNER JOIN ranking B ON A.item_code = B.item_code
WHERE B.main_category = '화장품/헤어'
"""
df4 = pd.read_sql(sql4, db)
df4

# 가격을 함수안에 넣어주고 따로 GROUP BY는 X
```

<div class="alert alert-block alert-warning">
<font color="blue" size="4em">연습문제5</font><br>
메인 카테고리별로 할인 가격이 10만원 이상인 상품이 몇개 있는지를 출력해보기
</div>

```python
sql5 = """ SELECT A.main_category, count(*) FROM ranking A
INNER JOIN items B ON A.item_code = B.item_code
WHERE dis_price >= 100000
GROUP BY A.main_category
"""
df5 = pd.read_sql(sql5, db)
df5
```

#### 서브쿼리를 이용하면

```python
sql5_1 = """ SELECT main_category, count(*) FROM ranking
WHERE item_code IN 
(SELECT item_code FROM items WHERE dis_price >= 100000)
GROUP BY main_category
"""
df5_1 = pd.read_sql(sql5_1, db)
df5_1
```

<div class="alert alert-block alert-warning">
<font color="blue" size="4em">다양한 복합쿼리 연습문제</font><br>
메인 카테고리, 서브 카테고리에 대해, 평균할인가격과 평균할인율을 출력해보기
</div>

```python
sql6 = """SELECT AVG(A.dis_price), AVG(A.discount_percent) FROM items A
INNER JOIN ranking B ON A.item_code = B.item_code
GROUP BY B.main_category, B.sub_category
"""
df6 = pd.read_sql(sql6, db)
df6
```

```python
sql6 = """SELECT B.main_category, B.sub_category, AVG(A.dis_price), AVG(A.discount_percent) FROM items A
INNER JOIN ranking B ON A.item_code = B.item_code
GROUP BY B.main_category, B.sub_category
ORDER BY B.main_category, B.sub_category
"""
df6 = pd.read_sql(sql6, db)
df6
# SELECT에 표기 필수
# ORDER BY 까지
```

<div class="alert alert-block alert-warning">
<font color="blue" size="4em">다양한 복합쿼리 연습문제</font><br>
판매자별, 베스트상품 갯수, 평균할인가격, 평균할인율을 베스트상품 갯수가 높은 순으로 출력해보기
</div>

```python
sql7 = """SELECT provider, count(*), AVG(dis_price), AVG(discount_percent)
FROM items
GROUP BY provider 
ORDER BY count(*) DESC
"""
df7 = pd.read_sql(sql7, db)
df7.head(20)
```

```python
sql7_1 = """SELECT A.provider, count(*), AVG(A.dis_price), AVG(A.discount_percent) FROM items A
INNER JOIN ranking B ON A.item_code = B.item_code
GROUP BY A.provider 
ORDER BY count(*) DESC
"""
df7_1 = pd.read_sql(sql7_1, db)
df7_1.head(20)
```

<div class="alert alert-block alert-warning">
<font color="blue" size="4em">다양한 복합쿼리 연습문제</font><br>
각 메인 카테고리에서 베스트 상품 갯수가 20개 이상인 판매자의 판매자별 평균할인가격, 평균할인율, 베스트 상품 갯수 출력해보기
</div>

```python
sql8 = """SELECT A.provider, count(*), AVG(A.dis_price), AVG(A.discount_percent) FROM items A
INNER JOIN ranking B ON A.item_code = B.item_code
GROUP BY A.provider HAVING count(*) >= 20
ORDER BY count(*) DESC
"""
df8 = pd.read_sql(sql8, db)
df8
```

```python
sql8_1 = """SELECT provider, count(*), AVG(dis_price), AVG(discount_percent) FROM items
WHERE item_code IN (SELECT item_code FROM ranking)
GROUP BY provider HAVING count(*) >= 20
ORDER BY count(*) DESC
"""
df8_1 = pd.read_sql(sql8_1, db)
df8_1
```
