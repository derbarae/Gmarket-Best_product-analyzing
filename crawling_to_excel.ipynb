### 2. 실전 예제: 크롤링 & 크롤링
* 이커머스(지마켓) 베스트100 상품 타이틀/가격, 그리고 판매업체 추출하기

```python
bestlists = soup.select('div.best-list')
bestitems = bestlists[1]
products = bestitems.select('ul > li')

for index, product in enumerate(products):
    item_seller=product.find('div', 'goods-view')
    print(index+1, item_seller.string)
#아니네
```

```python
bestlists = soup.select('div.best-list')
bestitems = bestlists[1]
products = bestitems.select('ul > li')

for index, product in enumerate(products):
    item_name=product.select_one('a.itemname')
    item_price=product.select_one('div.s-price')
    res=requests.get(item_name['href'])
#href 에 '' 처리
    soup=BeautifulSoup(res.content, 'html.parser')
    seller=soup.select_one('div.item-topinfo_headline > p > span.text__seller')
    print(index+1, seller.get_text())
```

### 3. 실전 예제: 크롤링 & 크롤링 + 엑셀
* 이커머스(지마켓) 베스트100 상품 타이틀/가격, 그리고 판매업체 추출하기

```python
import requests
from bs4 import BeautifulSoup
import openpyxl

res=requests.get('http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G06')
soup=BeautifulSoup(res.content, 'html.parser')
data=soup.select('div.best-list')[1]
products=data.select('ul > li')

excel_file=openpyxl.Workbook()
excel_sheet=excel_file.active
excel_sheet.column_dimensions['B'].width=100
excel_sheet.column_dimensions['C'].width=20
excel_sheet.column_dimensions['D'].width=20
excel_sheet.append(['랭킹','이름','가격','핀매자'])

for index, product in enumerate(products):
    name=product.select_one('a.itemname')
    price=product.select_one('div.item_price > div.s-price > strong')
    res2=requests.get(name['href'])
    soup2=BeautifulSoup(res2.content, 'html.parser')
    seller=soup2.select_one('div.item-topinfo_headline > p > span.text__seller')
    
    excel_sheet.append([index+1,name.get_text(),price.get_text(),seller.get_text()])
    
#여길까먹었어
excel_file.save('gmarket_best.xlsx')
excel_file.close()
```
